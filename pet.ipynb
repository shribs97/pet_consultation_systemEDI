```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
```

```python
# Load the dataset
data = pd.read_csv('pet3.csv')
```

```python
# Define the features and target
numerical_features = ['Age']
categorical_features = ['Species', 'Change in behaviour pattern', 'Description of behavior', 'Aggression', 'Compulsivebehaviour', 'Elimination','Disease']
target = 'Disease'
```

```python
# Remove rows with missing values
data = data.dropna()
```

```python
# One-hot encode the categorical features
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(data[categorical_features])
```

```python
# Combine the numerical and encoded categorical features
features = pd.concat([data[numerical_features], pd.DataFrame(encoded_data.toarray())], axis=1)
```

```python
# Convert the column names to strings
features.columns = features.columns.astype(str)
```

```python


# Split the data into training and testing sets
if len(features) != len(data[target]):
    print(f"features has length {len(features)} while data[target] has length {len(data[target])}")
else:
    X_train, X_test, y_train, y_test = train_test_split(features, data[target], test_size=0.2)

```

```python
# Convert the column names and target variable to strings
X_train.columns = X_train.columns.astype(str)
y_train = y_train.astype(str)
```

```python
# Create an instance of the DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
# Fit the model to the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python

# Collect the information from the user
age = input('Enter the age of the pet: ')
species = input('Enter the species of the pet (dog or cat): ')
change_in_behaviour = input('Has there been a change in the pet\'s behaviour? (yes or no): ')
description_of_behavior = input('Enter a description of the pet\'s behavior: ')
aggression = input('Is the pet showing aggression? (yes or no): ')
compulsivebehaviour = input('Is the pet exhibiting compulsive behavior? (yes or no): ')
elimination = input('Is the pet having elimination issues? (yes or no): ')
```

```python
# Create a feature vector for the new pet
new_data = pd.DataFrame({
    'Age': [age],
    'Species': [species],
    'Change in behaviour pattern': [change_in_behaviour],
    'Description of behavior': [description_of_behavior],
    'Aggression': [aggression],
    'Compulsivebehaviour': [compulsivebehaviour],
    'Elimination': [elimination]
})
```

```python
# Encode the categorical features
new_data_encoded = pd.get_dummies(new_data, columns=categorical_features)
```

```python
# Make sure the new feature vector has the same columns as the training data
new_data_encoded = new_data_encoded.reindex(columns=X_train.columns, fill_value=0)
```

```python
# Make a prediction for the new pet
prediction = clf.predict(new_data_encoded)
```

```python
# Print the predicted disease
print(f'Predicted disease: {prediction[0]}')
```

```python
# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
